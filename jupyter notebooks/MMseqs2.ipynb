# Searching

```bash
head QUERY.fasta
```

```bash
head SwissProt.fasta #targetDB
```

```bash
mkdir queryDB
mkdir targetDB
```

```bash
# convert your FASTA file into a sequence DB.
mmseqs createdb QUERY.fasta queryDB/queryDB
```

```bash
ls -l queryDB/
# The queryDB and queryDB.index files contain the amino acid sequences, 
# The queryDB_h and queryDB_h.index file contain the FASTA headers. 
# The queryDB.lookup file contains a list of tab separated fields that map from the internal identifier to the FASTA identifiers.
```

```bash
mmseqs createdb SwissProt.fasta targetDB/targetDB
```

```bash
ls -l targetDB/
```

```bash
#It is recommended to compute the index if the targetDB is reused for several searches.
mmseqs createindex targetDB/targetDB tmp
```

```bash
# MMseqs2 can produce a high IO on the file system. It is recommended to create a temporary folder on a local drive.
# mkdir tmp
```

```bash
mkdir resultDB
```

```bash
mmseqs search queryDB/queryDB targetDB/targetDB resultDB/resultDB tmp -a 

#-format-output "query,target,qaln,taln,alnlen,evalue,nident,pident,bits" not working?
```

```bash
mmseqs convertalis queryDB/queryDB targetDB/targetDB resultDB/resultDB resultDB/resultDB.m8
```
